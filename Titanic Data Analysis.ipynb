```python
titanic_data.isnull().sum()
```

```python
sex=pd.get_dummies(titanic_data['Sex'])['male']
sex
```

```python
embark=pd.get_dummies(titanic_data['Embarked'],drop_first=True)
embark
```

```python
pcl=pd.get_dummies(titanic_data['Pclass'],drop_first=True)
pcl
```

```python
titanic_data=pd.concat([titanic_data,sex,embark,pcl],axis=1)
```

```python
titanic_data
```

```python
titanic_data.drop(['Sex','Embarked','PassengerId','Name','Ticket'],axis=1,inplace=True)
titanic_data
```

```python
titanic_data.drop('Pclass',axis=1,inplace=True)
titanic_data
```

```python
age_mean=np.mean(titanic_data['Age'])
titanic_data['Age']-=age_mean
age_standard_deviation=np.std(titanic_data['Age'])
titanic_data['Age']/=age_standard_deviation
```

```python
fare_mean=np.mean(titanic_data['Fare'])
titanic_data['Fare']-=fare_mean
fare_standard_deviation=np.std(titanic_data['Fare'])
titanic_data['Fare']/=fare_standard_deviation
```

```python
titanic_data
```

```python
X=titanic_data.drop("Survived",axis=1)
y=titanic_data['Survived']
```

## Predicting using Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
# Confusion matrix is 2 x 2 matrix
#              | Predicted No   Predicted Yes
#  Actual No   |    105              21
# Actual Yes   |     25              63
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
